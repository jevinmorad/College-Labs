# Step 7: Split Data into Training and Testing Sets
The dataset is split into training and testing sets for model evaluation.

```python
x = df.drop('target',axis=1)
y = df['target']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=80)
```

# Step 8: Train Decision Tree
We train a Decision Tree Classifier on the training data. You have to also check for KNeighborsClassifier and  GaussianNB

```python

from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
```

```python
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, y_pred)
```

# Cross Validation

```python
from sklearn.model_selection import KFold
```

```python
k = KFold(n_splits=10)
```

```python
x
```

```python
for train_index, test_index in k.split(x):
    x_train, x_test, = x.values[train_index], x.values[test_index]
    y_train, y_test = y.values[train_index], y.values[test_index]
    model1 = DecisionTreeClassifier()
    model1.fit(x_train, y_train)
    print(model1.score(x_test, y_test))
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cross_val_score(model,x,y,cv=10)
```

# BS4 | web Scraping to get Data

```python
import requests
import bs4
```

```python
req = requests.get('https://darshan.ac.in/faculty-list/btech-computer')
```

```python
req.status_code
```

```python
soup = bs4.BeautifulSoup(req.text, "html.parser")
```

```python
data = soup.select("body > main > div:nth-child(5) > div > div > div > div > div")
```

```python
len(data)
```

```python
for f in data:
    name = f.select('h2')[0].get_text().strip()
    d1 = f.select('span')[0].text.strip()
    d2 = f.select('span')[1].text.strip()
    d3 = f.select('span')[2].text.strip()
    d4 = f.select('span')[3].text.strip()
    print(f'{name} ---> {d1}, {d2}, {d3} = {d4}')
```

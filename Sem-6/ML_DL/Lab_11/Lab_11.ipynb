```python
import numpy as np
```

```python
X_or = np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
])

X_or
```

```python
y_or = np.array([0,0,0,1])
y_or
```

```python
# weights = np.array([0.5,0.7])
# bias = 0.9
# lr = 0.1
# epochs = 5
```

```python
weights = np.array([0.0,0.0])
bias = 0.0
lr = 0.1
epochs = 5
```

```python
def step_function(x):
    if x >= 0:
        return 1
    else:
        return 0
```

```python
for epoch in range(epochs):
    print("Epoch:", epoch + 1)
    total_mistakes = 0

    for i in range(4):
        x = X_or[i]
        y = y_or[i]

        z = np.dot(x, weights) + bias
        predicted = step_function(z)

        error = y - predicted

        print("Predicted:", predicted, "Actual:", y)

        if error != 0:
            weights += lr * error * x
            bias += lr * error
            total_mistakes += 1

    if total_mistakes == 0:
        print("Correct on epoch", epoch + 1)
```

```python
def inference_XOR(x,y):
    z = (x * weights[0]) + (y * weights[1]) + bias
    predicted = step_function(z)
    return predicted

print("AND(0,0):", inference_XOR(0,0))
print("AND(0,1):", inference_XOR(0,1))
print("AND(1,0):", inference_XOR(1,0))
print("AND(1,1):", inference_XOR(1,1))
```

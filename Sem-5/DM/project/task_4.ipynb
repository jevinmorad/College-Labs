**Dataset**: https://www.kaggle.com/datasets/vijayuv/onlineretail



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Step 1: Load Data

```python
df = pd.read_csv('datasets/OnlineRetail.csv', encoding='ISO-8859-1')
df
```

# Step 2: Drop Missing Invoices or Descriptions

```python
df.dropna(inplace=True)
df
```

# Step 3: Remove Canceled Invoices <BR> Start with letter 'c'

```python
df = df[~df['InvoiceNo'].astype(str).str.startswith('C')]
df
```

# Step 4: Keep Only Positive Quantities

```python
df = df[df['Quantity'] > 0]
df
```

# Step 5: Standardize Item Descriptions | Lower case and remove space

```python
df['Description'] = df['Description'].str.lower()
df
```

# Step 6: Filter for a Single Country (e.g., United Kingdom)

```python
df_uk = df[df['Country'] == 'United Kingdom']
df_uk
```

# Step 7: Remove Duplicates

```python
df_uk = df_uk.drop_duplicates()
df_uk
```

# Step 8: Add TotalPrice column , TotalPrice = Quantity *  UnitPrice

```python
df_uk['TotalPrice'] = df_uk['Quantity'] * df_uk['UnitPrice']
df_uk
```

# Step 9 : Check for Top 10 Selling Products

```python
top_products = df_uk.groupby('Description')['TotalPrice'].sum().sort_values(ascending=False).head(10)
print(top_products)
```

# Step 10 : Identify Most Active Customers

```python
most_active_customer = df_uk['CustomerID'].value_counts().head(10)
print(f"Most active customer ID: {most_active_customer}")
```

# Step 11: Time-Based Analysis | Convert InvoiceDate to datetime and check popular months/hours.

```python
df_uk['InvoiceDate'] = pd.to_datetime(df_uk['InvoiceDate'])

df_uk['Month'] = df_uk['InvoiceDate'].dt.month
popular_months = df_uk['Month'].value_counts().sort_index()
plt.figure(figsize=(8,4))
popular_months.plot(kind='bar')
plt.title('Monthly Sales Volume')
plt.xlabel('Month')
plt.show()
```

# Step 12: Convert to Basket Format

```python
basket = df_uk.groupby(['InvoiceNo', 'Description'])['Quantity'].sum().unstack().fillna(0)
basket = basket.applymap(lambda x: 1 if x > 0 else 0)
basket
```

# Step:13 : Save Basket Format into CSV file

```python
basket.to_csv('basket_format.csv')
```
